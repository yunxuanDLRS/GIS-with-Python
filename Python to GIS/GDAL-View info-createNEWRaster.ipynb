```python
data1="/content/drive/MyDrive/AI data/DEM/ShiChuan.tiff"
data2='/content/drive/MyDrive/AI data/DEM/srtm_25_02.img'
```

```python
ds = gdal.Open(data2)

# basic raster info
#  查看size
print(ds.RasterXSize,ds.RasterYSize)
#  获得投影信息
print(ds.GetProjection())
#  地理仿射变换参数
print(ds.GetGeoTransform())
#  band num
print("num of raster band:",ds.RasterCount)
#  band info
band1 = ds.GetRasterBand(1)
print("No data value:",band1.GetNoDataValue())
print("min value:",band1.GetMinimum())
print("max value:",band1.GetMaximum())
print("data type:",band1.GetUnitType())
# trans to array
band_array = ds.GetRasterBand(1).ReadAsArray()
print("band_array.shape:",band_array.shape)
print("band_array.Max:",band_array.max())
print("band_array.Min:",band_array.min())
```

```python
# Create a Raster --- 1
# 会覆盖原来的文件

fn='/content/drive/MyDrive/AI data/DEM/srtm_25_02.tif'

driver_tiff = gdal.GetDriverByName("GTiff")

print("starting raster creation")
new_srtm_25_02_tif = driver_tiff.Create(fn,xsize=600,ysize=600,bands=1,eType=gdal.GDT_Float32)
print('raster created')
print(new_srtm_25_02_tif.RasterXSize,new_srtm_25_02_tif.RasterYSize)
#del(new_srtm_25_02_tif)
```

```python
# Create a Raster --- 2
# 会创建新的文件

# new raster filename
fn='new.tif'
# existing raster filename
fn2='/content/drive/MyDrive/AI data/DEM/srtm_25_02.tif'
# open existing raster
ds = gdal.Open(fn2)
print("ds.RasterXSize,ds.RasterYSize:",ds.RasterXSize,ds.RasterYSize)
# get the geotransform and projection of an existing rasyer
geot = ds.GetGeoTransform()
proj = ds.GetProjection()
# get a gdal driver to use for raster creation
driver_tiff = gdal.GetDriverByName("GTiff")
# create the new raster dataset
print("starting raster creation")
new_ds = driver_tiff.Create(fn,xsize=ds.RasterXSize,ysize=ds.RasterYSize,bands=1,eType=gdal.GDT_Float32)
# set the geotransform
new_ds.SetGeoTransform(geot)
# set the projection
new_ds.SetProjection(proj)
print('raster created')
print(new_ds.RasterXSize,new_ds.RasterYSize)
del(new_ds)
```
